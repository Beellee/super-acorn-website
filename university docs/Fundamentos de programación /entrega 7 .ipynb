Continuamos con el análisis de la demanda. Ahora haremos uso de técnicas de *machine learning* no supervisadas para buscar patrones de consumo. En concreto, intentaremos buscar **dos clústers**, uno característico del patrón de consumo en días laborables y otro, en días no laborables.

(a) A partir del dataframe resultante del ejercicio 1, extrae la serie correspondiente a la demanda del 2021 y modifica sus dimensiones según la tupla (365, 24). El objetivo es crear un `array` de `numpy` cuyas filas sean los días del año y cuyas columnas, las horas de cada día.  <span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span>  **(0.5 puntos)**

```python
import pandas as pd
import numpy as np

# Creamos una tabla con los dias del año como filas y las horas como columnas 
df["fecha"] = pd.to_datetime(df["fecha"])
table = df.pivot_table(values='Demanda programada PBF total', index=df['fecha'].dt.to_period('D'), columns=df['fecha'].dt.hour, aggfunc=np.mean)

# Convertimos la tabla a un array de numpy
demand_array = table.to_numpy()
demand_array
```

(b) Crea un modelo `KMeans`con **dos** clústers usando el algoritmo 'elkan' y traza la gráfica de los centroides de cada *clúster* en función de las horas del día.  <span style = "font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> **(1.0 punto)**

**Nota**: En el enlance siguiente encontarás más información sobre la función [KMeans](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html).

```python
# Respuesta
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# Para evitar que de error voy a reemplazar los valores nan por 0
nan_val = np.isnan(demand_array)
demand_array[nan_val] = np.nan_to_num(demand_array[nan_val],nan=0)

# Creo un modelo KMeans con dos clústers usando el algoritmo 'elkan' 
kmeans = KMeans(n_clusters=2, algorithm='elkan')
# Ajusto el modelo a los datos
kmeans.fit(demand_array)

# Obtengo los centroides de cada cluster
centroids = kmeans.cluster_centers_

# Trazo la gráfica
plt.plot(centroids.T)
plt.xlabel('Hora del día')
plt.ylabel('Demanda programada PBF total')
plt.legend(['Cluster 1', 'Cluster 2'])
plt.show()
```

(b) Con el *dataset* normalizado, crea un modelo de la clase [`PCA`](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html) sin especificar el número de componentes a retener y verifica que el número de componentes finales coincide con el de columnas del *dataset* de entrenamiento.  <span style="font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span>  **(0.5 puntos)**

```python
# Respuesta
from sklearn.decomposition import PCA

# creamos un modelo de la clase PCA sin especificar el número de componentes 
pca = PCA(n_components=None)
# ajustamos el modelo a los datos 
pca.fit(df_normalized)
# cogemos el numero de columnas en el dataset de entrenamiento 
num_columns = df_normalized.shape[1]
# cogemos el número de componentes finales 
num_components = len(pca.explained_variance_ratio_)

# vemos si el numero de componentes finales coincide con el de columnas del dataset de entrenamiento
if num_components == num_columns:
    print("El número de componentes finales coincide con el número de columnas en el conjunto de datos de entrenamiento.")
else:
    print("El número de componentes finales NO coincide con el número de columnas en el conjunto de datos de entrenamiento.")
```

(c) ¿Cuántas componentes son necesarias para explicar como mínimo el 80% de la varianza del conjunto de datos de entrenamiento? <span style = "font-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px;">EI</span> **(1.0 punto)**

```python
# Respuesta
import numpy as np
# seleccionamos el ratio de varianza explicada para cada componente 
explained_variance = pca.explained_variance_ratio_

# calculamos la varianza explicada acumulada 
cumulative_explained_variance = np.cumsum(explained_variance)

print(cumulative_explained_variance)

"""
Como podemos observar en el array "cumulative_explained_variance" necesitaríamos
10 componentes para poder explicar un 80.668697% de la varianza

También podríamos utilizar el siguiente código: 
"""
num_components = np.argmax(cumulative_explained_variance >= 0.8) + 1
print(f"\n Necesitamos {num_components} componentes para poder explicar al menos el 80% de la varianza")
```
