b) Calcula utilitzando las herramientas de NetworkX todas las posibles rutas entre `Philadelphia` y `San Francisco`, y muéstra la primera (la que cruza menos ciudades) y la última (la que cruza más ciudades). ¿Cuantas rutas aparecen? ¿Cuantas ciudades contiene la más larga ?

**NOTA** Puedes consultar en [shortest_simple_paths](https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.simple_paths.shortest_simple_paths.html) cómo generar caminos entre nodos.

<span style="font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span>
**(1 punto)**

````python
# Respuesta
# Seleccionamos todas las rutas posibles entre Philadelphia y San Francisco
all_paths = list(nx.all_simple_paths(G, "Philadelphia", "San Francisco"))
# Imprimimos el número de rutas
print("\n Numero de rutas:", len(all_paths))
# Ordenamos las rutas por el número de ciudades 
all_paths.sort(key=len)
# Seleccionamos la primera (con menos ciudades) y la última (con más)
shortest_path = all_paths[0]
longest_path = all_paths[-1]
# Imprimimos los resultados 
print("\n Ruta más corta:", shortest_path)
print("\n Ruta más larga:", longest_path)
# Imprimimos el nº de ciudades en la ruta más larga 
print("\n Número de ciudades en la ruta más larga:", len(longest_path))
````
