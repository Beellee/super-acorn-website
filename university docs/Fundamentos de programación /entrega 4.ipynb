### Ejercicio 4
Calcula la [media ponderada](https://es.wikipedia.org/wiki/Media_ponderada) de la duración de los viajes para el año 2021, cogiendo como *pesos* el número de turistas mensuales de ese mismo año.

**(1 punto)** $\color{red}{\text{EI}}$

```python
# Respuesta
import numpy as np

dataDur21= data.loc[(data['Concepto turístico'] == 'Duración media de los viajes') & (data['año'] == 2021)] #(data['Concepto turístico'] == 'Duración media de los viajes') & 
dataTour21= data.loc[(data['Concepto turístico'] == 'Turistas') & (data['año'] == 2021)]

val = dataDur21[['Total']]
wt = dataTour21[['Total']]

numerador = pd.DataFrame(wt.values * val.values)
sumaNumerador = numerador[0].sum()
sumaDenominador = wt['Total'].sum()
weigthedAv = round(sumaNumerador/sumaDenominador,2)
print(weigthedAv)

#Esto se puede hacer de una forma mucho más fácil y rápida usando la librería numpy
weigthedAv = round(np.average( dataDur21['Total'], weights = dataTour21['Total']),2)
print(weigthedAv)
```
