### Ejercicio 3
Cread el string **roxanneENNAXORanneRox** en una única expresión y utilizando solamente la variable m = `roxanne`.

**Nota**: Para resolver este ejercicio tenéis que utilizar las funciones para convertir un string en mayúsculas que podéis consultar en el seguiente [link](https://docs.python.org/3/library/stdtypes.html?highlight=upper#str.upper).
**(1 punto)** $\color{orange}{\text{EG}}$

```python
# Respuesta
m= "roxanne";
upper_m = m.upper(); #convertimos m a mayúsculas 
reversed_upper_m = upper_m [::-1]; #invertimos m (en mayúsculas(upper_m))
anne= m[3:7]; #seleccionamos los caracteres del 3 al 7(no incluido)
R= upper_m[0]; #seleccionamos la primera letra de upper_m
ox = m[1:3]; # seleccionamos los caracteres del 1 al 3(no incluido)

word = m + reversed_upper_m + anne + R + ox; # juntamos todas las variables en el orden que nos interesa en una variable llamada word

print(word); #imprimimos word en pantalla
```

### Ejercicio 5

Cread una lista de 5 elementos formada por números enteros diferentes. Mediante las funciones de la llibrería de Python `Numpy`, cread un código para responder cada una de las siguientes preguntas y mostrad el resultado por pantalla. 

a) Calculad la suma de todos los elementos de la lista.**(0.5 puntos)**

b) Calculad el máximo y el mínimo de la lista.**(0.5 puntos)**

c) Calculad la desviación estándard con dos números decimales.**(0.5 puntos)**

d) Ordenad la lista de mayor a menor.**Opcional**

**(1.5 punts)** $\color{green}{\text{NM}}$

```python
# Respuesta 
#importamos numpy como np
import numpy as np
lista =[-3, 5, 1, 8, 15]

# a)Suma de todos los elementos:
print("Suma: ", np.sum(lista));

# b)Máximo y mínimo de la lista:
print("Máximo: ", np.max(lista));
print("Mínimo: ", np.min(lista));

# c)Desviación estándard con dos números decimales:
print("Desviación estándar con dos decimales: ", round(np.std(lista), 2));

# d)Lista ordenada de mayor a menor:
print("Lista ordenada de mayor a menor: ", sorted(lista, reverse=True));
```
