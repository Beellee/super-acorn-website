(c) Para hacer un análisi rápido de las diferentes marcas, y precios de sus vehículos, queremos obtener los nombres de los diferentes fabricantes "Brands" y el precio medio de sus vehículos. Utilizando "groupby" crea un nuevo dataframe que tenga esta información y muéstrala. e) En este apartado vamos a generar 3 grupos con un número de muestras similares, utilizando el metodo "qcut". Etiquetaremos estos tres grupos en la columna "Price_Category", con las etiquetas "low","medium" y "high". Crea esta columna, y muestra cuántas filas hay para cada uno de sus valores, así como el precio medio de cada uno de estos grupos.
**(1 punto)** <span style = "font-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px;">EI</span>

```python
# Respuesta
df['Price_Category'] = pd.qcut(df['PriceEuro'], q=3, labels=['low', 'medium', 'high']) #creamos la columna y añadimos las etiquetas
groups = df.groupby('Price_Category')# agrupamos el dataframe según la columna Price_Category
print("Número de filas de cada grupo: \n",groups.size(), "\n") # mostramos el numero de filas de cada grupo
print("Media de cada grupo \n", groups['PriceEuro'].mean()) # mosrtramos la media de cada grupo 
```

(f)  Tras generar esta segmentación, los siguientes puntos que parecen importar a nuestro compañero son la aceleración y la autonomía. Crea un dataframe que contenga solo los vehículos de la "price_Category" "low" y normaliza las columnas "AccelSec" y "Range_Km". Normaliza dichas dos variables utilizando el módulo `MinMaxScaler` de sklearn. Como la Acceleración es mejor como más pequeña es, deberemos invertir los valores obteniendo un 0 para el valor de "AccelSec" mas alto, y un 1 para el "AccelSec" menor. Crea las dos columnas nuevas para los valores normalizados, y una tercera con el producto de los mismos, que llamaremos "Product".
**(1 punto)** <span style = "font-family: Courier New; background-color: #f2ae72; color: #000000; padding: 3px;">EI</span>

```python
# Respuesta
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()# Creamos el objeto MinMaxScaler 

# Seleccionamos las filas donde 'Price_Category' es "low"
df_low = df[df['Price_Category'] == 'low']

# Creamos dos nuevas columnas para los valores normalizados y normalizamos los valores 
df_low['AccelSec_Norm'] = scaler.fit_transform(df_low[['AccelSec']])
df_low['Range_Km_Norm'] = scaler.fit_transform(df_low[['Range_Km']])

#Invertimos los valores de 'AccelSec'
df_low['AccelSec_Norm'] = 1 - df_low['AccelSec_Norm']

# Creamos la columna 'Product' con el producto de los valores de las columnas 
df_low['Product'] = df_low['AccelSec_Norm'] * df_low['Range_Km_Norm'] 

#Podemos vel el dataframe resultante:
df_low
```

```python
# Respuesta

row_product = df_low['Product'].idxmax()# fila con el valor de 'Product' más alto 

# Print the values of the 'Model' and 'Product' columns of the selected row
print("El vehículo de la categoria de precios low que muestra el valor Product mas alto es: \n",df_low.loc[row_product, ['Model', 'Product']], "\n")

#Para saber si este coche es el más caro de los etiquetados con la actergoría low hacemos lo siguiente:
row_product_price = df_low.loc[row_product, 'PriceEuro'] #seleccionamos el valor de precio del vehiculo anterior 
row_price = df_low['PriceEuro'].max() #seleccionamos el valor de precio más alto 

print("el precio más alto de la categoría low es: ",row_price, "y el precio del vehiculo anterior es: ", row_product_price)
#como podemos comprobar son distintos, por lo tanto el coche anterior no es el más caro de los etiquetados en la categoría "low"
```
