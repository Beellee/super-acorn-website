a) Queremos hacer un programa que vaya sumando el dinero que gastamos y que, cuando lleguemos a un límite que definimos nosotros, nos avise y detenga el programa. El código debe tener las siguientes características:

- En cada iteración, incrementaremos el dinero que hemos gastado por un valor random entre 0 y 10 mediante la función [randint](https://docs.python.org/3/library/random.html#random.randint). <span style = "font-family: Courier New; background-color: #ffcc5c; color: #000000; padding: 3px; ">EG</span> (**0.5 puntos**)
- En cada iteración, se tiene que mostrar por pantalla cuánto dinero hemos gastado y cuánto nos quedan. b) El código anterior es muy interesante y queremos utilizarlo para apuntar lo que gastamos cuando vamos a comprar. Por este motivo, es conveniente ponerlo en forma de función, ya que nos permitirá interaccionar con él más fácilmente.
Para ello, tendremos que tener en cuenta los siguientes puntos:
- La función deberá tener como parámetro de entrada el límite máximo que podemos gastar. <span style = "font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> (**0.25 puntos**)
- El usuario deberá introducir cada uno de los gastos mediante la función **input**, ya no será un valor aleatorio. <span style = "font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> (**0.5 puntos**)
- Tal y como se ha comentado en el apartado anterior, se deberá mostrar por pantalla cuánto dinero hemos gastado, lo que nos queda, y, si llegamos al límite, se deberá avisar al usuario. <span style = "font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> (**0.75 puntos**)

(**1.5 puntos**)

```python
## Respuesta

def compra(max):
  dinero_gastado = 0 # creamos una variable donde guardaremos el dinero que vayamos gastando 

  while dinero_gastado<max: # mientras el dinero gastado sea menor que el límite que hemos establecido ejecutamos el siguiente código:

    add = float(input("¿cuánto acabas de gastar? ")) #preguntamos al usuario cuánto dinero ha gastado y lo almacenamos como número (que puede tener decimales)
    
    dinero_gastado += add # lo añadimos al dinero que llevamos gastado 

    print(f"has gastado un total de {dinero_gastado}, te quedan {round(max-dinero_gastado, 2)}") #imprimimos por pantalla cuanto dinero nos hemos gastado y cuanto dinero queda 
    if dinero_gastado >= max: # si el dinero gastado es mayor o igual que el máximo establecido: 
      print(f"has alcanzado el límite de dinero que podías gastar, te has pasado del límite de {max} por {round(dinero_gastado-max, 2)}") # imprimimos un aviso de que hemos alcanzado el límite y decimos por cuánto nos hemos pasado (usamos la función round para redondear a dos decimales y evitar tener resultados como 0.10000000000000009) 
      
compra(20)   
```

c) La calculadora de gastos funciona bien, pero nos hemos dado cuenta de que si el usuario introduce por error algo que no sea un número, todo el programa falla. Para evitar esta situación, utiliza tu conocimiento sobre gestión de errores y excepciones para obtener el siguiente resultado:
- Cuando el usuario introduzca un valor erróneo, el programa le avisará, en lugar de fallar directamente. <span style = "font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> (**0.5 puntos**)
- Una vez que el usuario ha sido avisado, el programa permite introducir el valor de nuevo una vez más. Si vuelve a introducir un valor erróneo, el programa se cierra. <span style = "font-family: Courier New; background-color: #82b74b; color: #000000; padding: 3px; ">NM</span> (**0.5 puntos**)

(**1 punto**)

```python
## Respuesta
def compra(max):
  dinero_gastado = 0 # creamos una variable donde guardaremos el dinero que vayamos gastando 
  fallos = 0 # creamos una variable para guardar el número de fallos que comete el usuario 

  while dinero_gastado<max : # mientras el dinero gastado sea menor que el límite que hemos establecido ejecutamos el siguiente código:

    try: # usamos try para ejecutar el cógigo 
      if fallos == 2: # si el número de fallos es igual a dos el programa se cierra 
        break
      # si el número de fallos es inferior a 2 se ejecuta el código siguiente:
      add = float(input("¿cuánto acabas de gastar? ")) #preguntamos al usuario cuánto dinero ha gastado y lo almacenamos como número (que puede tener decimales)

      dinero_gastado += add # lo añadimos al dinero que llevamos gastado 

      print(f"has gastado {dinero_gastado} euros, te quedan {round(max-dinero_gastado, 2)}") #imprimimos por pantalla cuanto dinero nos hemos gastado y cuanto dinero queda 
      
      if dinero_gastado >= max: # si el dinero gastado es mayor o igual que el máximo establecido: 
        print(f"has alcanzado el límite de dinero que podías gastar, te has pasado del límite de {max} por {round(dinero_gastado-max, 2)}") # imprimimos un aviso de que hemos alcanzado el límite y decimos por cuánto nos hemos pasado (usamos la función round para redondear a dos decimales y evitar tener resultados como 0.10000000000000009) 
    
    except ValueError: # si se produce un ValueError (el usuario introduce un valor erróneo)
      fallos += 1 # contabilizamos que se ha producido un fallo 
      print("debes introducir un número con separador decimal (.) ") # avisamos al usuario 

compra(2)   
```
